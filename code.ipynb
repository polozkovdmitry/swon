```python
%%manim -qh -v  WARNING Elastance

from manim.utils.color import Colors

class Elastance(MovingCameraScene):
    def construct(self):

      ### title, axes ###

        # title = Title("Mamikon application")
        # self.play(FadeIn(title))

        axes = Axes(
            x_range=[-3, 1, 0.25],
            y_range=[-0.5, 2, 0.25],
            tips=False,
            axis_config={'tick_size' : 0.05,
                         "include_numbers": False})
        # axes.move_to(ORIGIN).to_edge(DOWN)
        axes.move_to(ORIGIN)


        x_label = MathTex(r"t")
        x_label.next_to(axes.get_x_axis().get_corner(UR), UL)

        y_label = MathTex(r"\text{exp}\{t\}", color=BLUE)
        y_label.next_to(axes.get_y_axis().get_corner(UR), RIGHT)

        origin_label = MathTex(r"t_0")
        origin_label.next_to(axes.coords_to_point(0, 0), DR)

        ### graph ###

        s_shape_graph = axes.plot(
            lambda x: np.exp(x),
            color=BLUE,
            x_range=[-3, np.log(2)])

        self.play(FadeIn(axes))

        self.play(
            FadeIn(x_label),
            FadeIn(y_label),
            FadeIn(origin_label)
        )

        self.play(Create(s_shape_graph)) # THIS IS THE EXPONENTIAL FUNCTION WE WANT TO INTEGRATE
        axes.add(s_shape_graph)


        ### ??? ###

        # alpha = ValueTracker(0.01)  # idk if this is needed atm

        # # function for drawing the tangent line
        # draw_tangent = (lambda:
        #     TangentLine(s_shape_graph, alpha.get_value(), 10, color=YELLOW))

        # # always redraw the tangent line, i.e. update when alpha changes
        # tangent = always_redraw(draw_tangent)

        # self.play(FadeIn(tangent))
        # self.play(alpha.animate.set_value(1), run_time=2)
        # self.play(FadeOut(tangent))

        txt = Tex("Lets draw a tangent at $t_0$", font_size=60).to_edge(UL, buff=LARGE_BUFF)
        # txt.add_background_rectangle(opacity=0.2, color=BLUE, buff=0.25)
        # txt.next_to(y_label, LEFT, buff=LARGE_BUFF)

        self.play(Write(txt), FocusOn(Dot(axes.coords_to_point(0, 1)), run_time=3), time=3) # THIS IS THE INTERSECTION POINT
        self.play(FadeOut(txt))


        beta = ValueTracker(0)

        dot_l_for_tangent = always_redraw(lambda: Dot(axes.coords_to_point(beta.get_value() - 1, 0),
                                                         color=Colors.maroon_a.value))
        dot_u_for_tangent = always_redraw(lambda: Dot(axes.coords_to_point(beta.get_value(), np.exp(beta.get_value())),
                                                          color=Colors.maroon_a.value))
        draw_tangent_small = (lambda:
                                  axes.plot(
                                      lambda x: np.exp(beta.get_value()) * x + np.exp(beta.get_value()) * (1 - beta.get_value()),
                                      color=Colors.maroon_d.value,
                                      x_range=[beta.get_value() - 1, beta.get_value()]
                                      )
                                  )
        tangent_small = always_redraw(draw_tangent_small)

        draw_tangent_full_dashed = (lambda: ##
                                 DashedVMobject(
                                     axes.plot(
                                         lambda x: np.exp(beta.get_value()) * x + np.exp(beta.get_value()) * (1 - beta.get_value()),
                                         color=Colors.maroon_a.value,
                                         x_range=[beta.get_value() - 1 - 3, beta.get_value() + 5],
                                         )
                                     )
                                 )
        tangent_full_dashed = always_redraw(draw_tangent_full_dashed)
        vertical_dashed = always_redraw(lambda: DashedVMobject(Line(axes.coords_to_point(beta.get_value(), np.exp(beta.get_value())),
                                                                    axes.coords_to_point(beta.get_value(), 0),
                                                                    color=GRAY)))
        vertical_dashed.set_z_index(axes.z_index - 1)
        tangent_vg = VGroup(dot_l_for_tangent,
                                dot_u_for_tangent,
                                tangent_small.set_z_index(dot_l_for_tangent.z_index - 1),
                                tangent_full_dashed.set_z_index(dot_l_for_tangent.z_index - 2),
                                vertical_dashed)

        self.play(Write(tangent_vg))

        ### braces ###

        # self.camera.frame.save_state()

        # self.play(self.camera.frame.animate.set(width=tangent_t_0_small.width * 5).move_to(tangent_t_0_small))
        # s_shape_graph.animate.set_opacity(0.5))

        b_d = Brace(tangent_small, color=GREEN, buff=0.2, direction = DOWN).add_updater(lambda x: x.next_to(tangent_small, DOWN,  buff=0.2))
        b_d_label = Tex("$B$", color=GREEN, font_size=34)
        b_d_label.next_to(b_d, DOWN, buff=0.15)
        b_r = Brace(tangent_small, color=GREEN, buff=0.2, direction = RIGHT)
        b_r_label = Tex("$A$", color=GREEN, font_size=34)
        b_r_label.next_to(b_r, RIGHT, buff=0.15)

        from numpy import arctan, exp

        arc = Arc(angle=arctan(exp(0)) * 0.92,
                  radius=0.6,
                  arc_center=dot_l_for_tangent.get_center(), color=GREEN).set_z_index(tangent_small.z_index - 1)
        arc_label = Tex(r"$\alpha$", color=GREEN, font_size=34)
        arc_label.next_to(arc, RIGHT, aligned_edge=UP, buff=0.15)

        Braces = VGroup(b_r, b_r_label, b_d, b_d_label, arc, arc_label)
        self.play(FadeIn(Braces[:2])) # LET'S DETERMINE THE LENGTH ON THIS SEGMENT A
        self.play(FadeIn(Braces[2:4])) # AND THIS B ONE, TOO
        self.play(FadeIn(Braces[4:])) # SINCE WE HAVE A TANGENT LINE, WE MUST KNOW SOMETHING ABOUT THIS ANGLE ALPHA

        txt1 = Tex(r"$\text{tg}{ (\alpha) } = \frac{de^t}{dt}\right|_{t= {t_0} }$", font_size=50)
        txt2 = Tex(r"$\text{tg}{ (\alpha) } = e^t\right|_{ {t_0} }$", font_size=50)
        txt3 = Tex(r"$\text{tg}{ (\alpha) } = e^{ {t_0} }$", font_size=50)
        txtTg = VGroup(txt1, txt2, txt3).arrange(DOWN, center=False, aligned_edge=LEFT).to_edge(UL, buff=LARGE_BUFF)
        b_r_label_new = Tex(r"$e^{t_0}$",  color=GREEN, font_size=44).move_to(b_r_label)
        b_d_label_new = Tex(r"$1$",  color=GREEN, font_size=44).next_to(b_d, DOWN, buff=0.15)
        b_d_label_new.add_updater(lambda x: x.next_to(b_d, DOWN, buff=0.15))
        self.play(FadeIn(txtTg[0], shift=0.5*DOWN))
        self.play(TransformMatchingShapes(txtTg[0].copy(), txtTg[1]), run_time=1)
        self.play(TransformMatchingShapes(txtTg[1].copy(), txtTg[2]), run_time=1)
        self.play(TransformMatchingShapes(b_r_label, b_r_label_new), run_time=1)
        self.play(TransformMatchingShapes(b_d_label, b_d_label_new), run_time=1)
        self.play(FocusOn(b_d_label_new))
        self.play(FadeOut(txtTg))


        # self.play(Restore(self.camera.frame))

        # integral evaluation ###

        disappear = [arc, arc_label]
        self.play(FadeOut(*disappear))

        origin_label_0 = Tex(r"0")
        origin_label_0.next_to(axes.coords_to_point(0, 0), DR)
        b_r_label_0 = Tex(r"$1$",  color=GREEN, font_size=44).move_to(b_r_label)
        self.play(TransformMatchingShapes(origin_label, origin_label_0))
        self.play(FocusOn(origin_label_0))
        self.play(TransformMatchingShapes(b_r_label_new, b_r_label_0))
        self.play(FocusOn(b_r_label_0))


        textI = Tex(r"$$\int_{-\infty}^0 e^xdx$$", font_size=40)
        # textI.move_to(y_label, LEFT, buff=2 * LARGE_BUFF)
        textI.to_edge(UL, buff=LARGE_BUFF)

        areaMain = axes.get_area(
            s_shape_graph,
            x_range=(-3, 0),
            color=YELLOW,
            opacity=0.5)

        areaTriu = axes.get_area(
            tangent_small,
            x_range=(-1, 0),
            color=YELLOW,
            opacity=0.5)

        piecewise = axes.plot(lambda x: x + 1 if x > -1 else 0)

        areaOther = axes.get_area(
            graph=s_shape_graph,
            bounded_graph=piecewise,
            x_range=(-3, 0),
            color=YELLOW,
            opacity=0.5)

        # NOW WE'RE READY TO EVALUATE THE INTEGRAL USING MAMIKON TRICK
        self.play(FadeIn(areaMain), Write(textI), time=2) # THIS IS THE AREA WE SEEK TO FIND: IT STARTS ALL THE WAY DOWN AT NEGATIVE INFINITY AND ENDS AT ZERO
        self.wait(3)

        plus2 = Tex(r"$+$", font_size=35)
        equals2 = Tex(r"$=$", font_size=35)

        targetMainPseudo = areaMain.copy().scale(0.3)
        targetTriuPseudo = areaTriu.copy().scale(0.3)
        targetOtherPseudo = areaOther.copy().scale(0.3)

        arragement = Group(targetMainPseudo, equals2, targetOtherPseudo, plus2, targetTriuPseudo).arrange(direction=RIGHT, buff=0.5)
        arragement.next_to(textI, DOWN).align_to(textI, LEFT)
        equals1 = Tex(r"$=$", font_size=35).move_to(textI.get_right()).align_to(equals2, LEFT)

        areaMain.generate_target()
        areaTriu.generate_target()
        areaOther.generate_target()

        areaMain.target.set_fill(opacity=0.5).scale(0.3).move_to(arragement[0])
        areaOther.target.set_fill(opacity=0.5).scale(0.3).move_to(arragement[2])
        areaTriu.target.set_fill(opacity=0.5).scale(0.3).move_to(arragement[4])


        self.play(MoveToTarget(areaMain), # NOTICE THAT THIS AREA CAN BE SPLIT APART INTO TWO SHAPES
                  MoveToTarget(areaTriu),
                  MoveToTarget(areaOther),
                  FadeIn(plus2, equals1, equals2))

        plus1 = Tex(r"$+$", font_size=35).move_to(textI.get_right()).align_to(plus2, LEFT)
        onehalf = Tex(r"$$\frac{1}{2}$$", font_size=35).move_to(areaTriu.get_center()).set_y(equals1.get_center()[1])

        # self.wait(3) # AS WE'VE FIGURED OUT, THESE (highlight simultaneously with two separate colors the coresponding seqments of a triangle) SEGMENTS BOTH EQUAL TO 1. SO THE AREA MUST BE 1/2

        self.play(FadeIn(plus1, onehalf))

        # A/B somewhere

        ### triangle main ###

        self.play(beta.animate.set_value(0.5),
                  FadeOut(b_r, b_r_label_0), run_time=1.5)
        self.play(beta.animate.set_value(-2), run_time=2.5)
        self.play(beta.animate.set_value(0),
                  FadeIn(b_r, b_r_label_0), run_time=2)

        ### triangles other ###

        # for mamikon we must have a tangent line that's sliding over some shape
        # we have a shape -- a graph
        self.play(FocusOn(dot_u_for_tangent)) # and where does the rangent line touch?
        self.play(FocusOn(dot_l_for_tangent)) # where is the endpoint of a tangent line?

        # let's try to figure out what this
        self.play(FocusOn(areaOther))
        # portion equals to

        # let's trace it by sliding the tangent line

        # self.play(FadeOut(tangent_vg[3:])) # always except dots and lines

        dot_l_for_tangent = always_redraw(lambda: Dot(axes.coords_to_point(beta.get_value() - 1, 0),
                                                         color=Colors.maroon_a.value))
        dot_u_for_tangent = always_redraw(lambda: Dot(axes.coords_to_point(beta.get_value(), np.exp(beta.get_value())),
                                                          color=Colors.maroon_a.value))
        draw_tangent_small = (lambda:
                                  axes.plot(
                                      lambda x: np.exp(beta.get_value()) * x + np.exp(beta.get_value()) * (1 - beta.get_value()),
                                      color=Colors.maroon_d.value,
                                      x_range=[beta.get_value() - 1, beta.get_value()]
                                      )
                                  )
        tangent_small = always_redraw(draw_tangent_small)

        draw_tangent_full_dashed = (lambda:
                                 DashedVMobject(
                                     axes.plot(
                                         lambda x: np.exp(beta.get_value()) * x + np.exp(beta.get_value()) * (1 - beta.get_value()),
                                         color=Colors.maroon_a.value,
                                         x_range=[beta.get_value() - 1 - 3, beta.get_value() + 5],
                                         )
                                     )
                                 )
        tangent_full_dashed = always_redraw(draw_tangent_full_dashed)

        # def upper_graph(x):
        #   tangent = np.exp(beta.get_value()) * x + np.exp(beta.get_value()) * (1 - beta.get_value())
        #   s_shape_graph = np.exp(x)
        #   if x < beta.get_value():
        #     return s_shape_graph(x)
        #   else:
        #     return tangent(x)

        draw_area_other_dymanic = (lambda:
                                   axes.get_area(
                                       graph=axes.plot(lambda x: np.exp(beta.get_value()) * x + np.exp(beta.get_value()) * (1 - beta.get_value()) \
                                                       if x < beta.get_value() else np.exp(x)),
                                                 bounded_graph=piecewise,
                                                 x_range=(beta.get_value() - 1, 0),
                                                 color=Colors.maroon_a.value,
                                                 opacity=0.5
                                                 )
                                   )
        area_other_dymanic = always_redraw(draw_area_other_dymanic)

        tangent_vg_new = VGroup(dot_l_for_tangent,
                                dot_u_for_tangent,
                                tangent_small.set_z_index(dot_l_for_tangent.z_index - 1),
                                tangent_full_dashed.set_z_index(dot_l_for_tangent.z_index - 2),
                                area_other_dymanic)
        self.remove(tangent_vg, b_r_label_new, b_d_label_new)
        self.add(tangent_vg_new)


        self.play(beta.animate.set_value(-3), run_time=4.5)


        self.wait(4)

        # code up to here has a rendered video: https://drive.google.com/file/d/1o0Q9PkbwkXaGB7GKF_-5It5MkZSC5Cvi/view?usp=sharing
```
